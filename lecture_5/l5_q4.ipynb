```python
# -----------------------------------------------------------------------------------------------------
#   l1-q4, solving second order ODE
#   Author: Yi-Ming Ding
#   Updated: Mar 19, 2025
#   Reference: https://docs.scipy.org/doc/scipy/reference/generated/scipy.integrate.solve_ivp.html
# -----------------------------------------------------------------------------------------------------
```

For
$$
y''-2y'+2y=e^{2x}\sin(x)
$$
with $y(0)=-0.4$, and $y'(0)=-0.6$, by defining $y_1=y$ and $y_2=y'$, we reformulate it as
$$
\begin{split}
&y_1'=y_2, y_1(0)=-0.4  \\
&y_2'=2y_2-2y_1+e^{2x}\sin x, y_2(0)=-0.6
\end{split}
$$
then we can do similar thing as that in the last problem.

```python
import numpy as np
from scipy.integrate import solve_ivp

def equation(x, y):
    y1, y2 = y
    dy_dt = [y2, 2 * y2 - 2 * y1 + np.exp(2 * x) * np.sin(x)]
    return dy_dt

y0 = [-0.4, -0.6]   # the initial value

x_span = (0, 1)
x_eval = np.linspace(0, 1, 100)

sol = solve_ivp(equation, x_span, y0, t_eval=x_eval)
```

```python
import matplotlib.pyplot as plt

plt.plot(sol.t, sol.y[0], label="y(x)")
plt.plot(sol.t, sol.y[1], label="y'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
